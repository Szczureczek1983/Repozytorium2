```python
# importujemy biblioteki
import numpy as np
import pylab
import matplotlib.pyplot as plt
import matplotlib as mpl
%matplotlib inline
mpl.rcParams['figure.dpi'] = 100 # poprawia jakość wykresu
# podajemy parametry a,b i c
a = float(input("Podaj współczynnik a: "))
b = float(input("Podaj współczynnik b: "))
c = float(input("Podaj współczynnik c: "))
d = float((input("Podaj współczynnik d: ")))
X = pylab.arange(-100,110,10) 
fx = []
for i in X:
  fx = a*X**3+b*X**2+2*c*X + 3*d
# kreślimy wykres
plt.plot(X,fx, "-o", color="r")
plt.xlabel("Wartości X")
plt.ylabel("Wartości funkcji fx")
plt.title('Wykres funkcji fx = a*X$^{3}$+b*X$^{2}$+2*c*X+3*d')
plt.grid(True)
plt.show()
```
