```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
data = pd.read_csv('/content/1.01.+Simple+linear+regression.csv')
```

```python
data.head()
```

```python
data.describe()
```

Tworzymy pierwszą regresję linową

```python
y = data['GPA']
x1 = data['SAT']
```

```python
plt.scatter(x1,y)
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.show()
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y, x).fit()
results.summary()
```

```python
plt.scatter(x1,y)
y_hat = 0.0017*x1 + 0.275
fig = plt.plot(x1, y_hat, lw = 4, c='orange', label = 'regression line' )
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.legend()
plt.show()
```

```python
raw_data = pd.read_csv('/content/1.03.+Dummies.csv')
```

```python
data = raw_data.copy()
```

```python
data.head()
```

```python
data['Attendance'] = data['Attendance'].map({'Yes':1, "No":0})
```

```python
data.head()
```

```python
data.describe()
```

```python
y = data['GPA']
x1 = data[['SAT', 'Attendance']]
```

```python
x = sm.add_constant(x1)
results = sm.OLS(y,x).fit()
results.summary()
```

```python
plt.scatter(data['SAT'], y, c=data['Attendance'], cmap = 'RdYlGn_r')
yhat_no = 0.6439+0.0014*data["SAT"]
yhat_yes = 0.8665 + 0.0014*data['SAT']
y_hat = 0.0017*data['SAT'] + 0.275
fig = plt.plot(data['SAT'], y_hat, lw = 2, c='orange', label = 'regression line' )
fig = plt.plot(data['SAT'], yhat_no, lw = 2, c = 'green', label = ' Not attended')
fig = plt.plot(data['SAT'], yhat_yes, lw = 2, c = 'red', label= 'Attended')
plt.xlabel('SAT', fontsize = 20)
plt.ylabel('GPA', fontsize = 20)
plt.legend()
plt.show()
```

predykcje

```python
new_data = pd.DataFrame({'const':1, 'SAT':[1700, 1670], 
                         'Attendance': [0,1]})
new_data = new_data[['const', 'SAT', 'Attendance']]
new_data.rename(index = {0: 'Bob', 1: "Alce"})
new_data
```

```python
pred = results.predict(new_data)
pred
```

```python
pred_df = pd.DataFrame({'Predictions':pred})
joined = new_data.join(pred_df)
joined.rename(index={0:'Bob', 1:'Alice'})
```
