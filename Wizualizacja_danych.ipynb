```python
temp.head()
```

```python
bangkok=temp[temp['City'] == 'Bangkok']
bangkok.head()
```

```python
merged = pd.merge(co2, bangkok, how = 'inner', on = 'Date')
```

```python
merged.head()
```

```python
fig, ax=plt.subplots()
ax.plot(merged.index, merged['Interpolated'])
ax.plot(merged.index, merged['AverageTemperature'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2/Temperature')
plt.show()
```

```python
fig, ax=plt.subplots(figsize = (9,6))
ax.plot(merged.index, merged['Interpolated'], color = 'b')
ax.set_xlabel('Time')
ax.set_ylabel('CO2', color = 'b')
ax.tick_params('y', colors = 'b')
ax2=ax.twinx()
ax2.plot(merged.index, merged['AverageTemperature'], color = 'r')
ax2.set_ylabel('Temperature', color = 'r')
ax2.tick_params('y', colors = 'r')
```

```python
merged_sub=merged['2000':]
fig, ax=plt.subplots(figsize = (9,6))
ax.plot(merged_sub.index, merged_sub['Interpolated'], color = 'b')
ax.set_xlabel('Time')
ax.set_ylabel('CO2', color = 'b')
ax.tick_params('y', colors = 'b')
ax2=ax.twinx()
ax2.plot(merged_sub.index, merged_sub['AverageTemperature'], color = 'r')
ax2.set_ylabel('Temperature', color = 'r')
ax2.tick_params('y', colors = 'r')
```

```python
def plot_timestories(axis, x, y, color_def, xlab, ylab):
  axis.plot(x,y, color = color_def)
  axis.set_xlabel(xlab)
  axis.set_ylabel(ylab)
  axis.tick_params('y', color=color_def)
```

```python
fig, ax=plt.subplots(figsize = (9,6))
plot_timestories(ax, merged_sub.index, merged_sub['Interpolated'], 'b', 'Time', 'CO2')
ax2=ax.twinx()
plot_timestories(ax2, merged_sub.index, merged_sub['AverageTemperature'], 'r', 'Time', 'Temperature')
```

```python
pop=pd.read_csv('ML-datasets/population.csv', index_col = "City")
pop.head()
```

```python
groupby = pop.groupby('Sex')
both, female, male = [groupby.get_group(x) for x in groupby.groups]
```

```python
both2=both.sort_values('Value', ascending=False)
fig, ax=plt.subplots(figsize = (10,8))
ax.bar(both2.index, both2['Value'])
ax.xaxis.set_tick_params(rotation = 90)
ax.set_ylabel('Population')
plt.show()
```

```python
both2=both.sort_values('Value', ascending=False).head(10)
fig, ax=plt.subplots(figsize = (10,8))
ax.bar(both2.index, both2['Value'])
ax.xaxis.set_tick_params(rotation = 90)
ax.set_ylabel('Population')
plt.show()
```

```python
fig, ax=plt.subplots(figsize = (10,8))
ax.bar(male.index, male["Value"],label = 'Male')
ax.bar(female.index, female['Value'], bottom = male['Value'], label = 'Female')
ax.xaxis.set_tick_params(rotation = 90)
ax.set_ylabel('Population')
ax.legend()
plt.show()
```

```python
medals = pd.read_csv('ML-datasets/olympic.csv', index_col = "Year")
```

```python
medals.head()
```

```python
set(medals['Sport'])
```

```python
def groupby_column(dataframe, column):
  groupby = dataframe.groupby(column)
  return {x: groupby.get_group(x) for x in groupby.groups}
```

```python
all_sports = groupby_column(medals, 'Sport')
```

```python
fig, ax=plt.subplots(figsize = (10,8))
ax.bar("Football", all_sports['Football']['Height'].mean(), yerr =all_sports['Football']['Height'].std() )
ax.bar("Basketball",all_sports['Basketball']['Height'].mean(), yerr =all_sports['Basketball']['Height'].std())
ax.set_ylabel('Height [cm]', fontsize = 12)
ax.set_title('Wzrost zawodników', fontsize = 18)
plt.show()
```

```python
fig, ax=plt.subplots(figsize = (10,8))
ax.hist(all_sports['Football']['Height'], label = "Football", bins = 30)
ax.hist(all_sports['Basketball']['Height'], label = "Basketball", bins = 30, histtype = 'step')
ax.set_xlabel('Height [cm')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

```python
fig, ax=plt.subplots(figsize = (10,8))
ax.hist(all_sports['Football']['Height'], label = "Football", bins =50)
ax.hist(all_sports['Basketball']['Height'], label = "Basketball", bins = 50, histtype = 'step')
ax.set_xlabel('Height [cm')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

```python
fig, ax=plt.subplots(figsize = (10,8))
ax.hist(all_sports['Judo']['Weight'], label = "Judo", bins =50)
ax.hist(all_sports['Boxing']['Weight'], label = "Boxing", bins = 50, histtype = 'step')
ax.set_xlabel('weight [kg]')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

```python
fig, ax=plt.subplots(figsize = (10,8))
ax.boxplot([all_sports['Football']['Height'].dropna(), all_sports['Basketball']['Height'].dropna()])
ax.set_xticklabels(['Football', 'Basketball'])
ax.set_ylabel('Height [cm]')
plt.show()
```

```python
fig, ax=plt.subplots(figsize = (10,8))
ax.boxplot([all_sports['Judo']['Weight'].dropna(), all_sports['Boxing']['Weight'].dropna()])
ax.set_xticklabels(['Judo', 'Boxing'])
ax.set_ylabel('Weight [kg]')
plt.show()
```
